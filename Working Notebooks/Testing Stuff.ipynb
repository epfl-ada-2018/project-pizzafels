```python
df = pd.DataFrame({'id':[],'localisation':[],'roadtype':[],'severitycategory':[],'accidentyear':[],'accidenttype':[],\
                   'fsocommunecode':[],'accidenttypecode':[],'canton':[],'accidentday':[],'accidentmonth':[],'accidenthour':[]})
again = True
offset = 0
while again:
    print('Running for {} to {}'.format(offset,offset+200))
    url = '''
    https://api3.geo.admin.ch/rest/services/all/MapServer/identify?geometry=2455375,1065350,2872025,1329250&geometryFormat=geojson&geometryType=esriGeometryEnvelope&imageDisplay=1079,938,96&lang=fr&layers=all:ch.astra.unfaelle-personenschaeden_alle&mapExtent=2234169.4811259033,818419.8492296317,2935519.4811259033,1428119.8492296317&offset={}&returnGeometry=true&sr=2056&tolerance=0
    '''.format(offset)
    r = requests.get(url)
    a = 0
    for accident in r.json()['results']:
        id_ = accident['id']
        localisation = accident['geometry']['coordinates'][0]
        roadtype = accident['properties']['roadtype_fr']
        severitycategory = accident['properties']['severitycategory_fr']
        date = accident['properties']['accidentday_fr'].split(' / ')
        accidentyear = date[2].split(' ')[1]
        accidenttype = accident['properties']['accidenttype_fr']
        fsocommunecode = accident['properties']['fsocommunecode']
        accidenttypecode = accident['properties']['accidenttypecode']
        canton = accident['properties']['canton']
        accidentday = date[0]
        accidentmonth = date[2].split(' ')[0]
        accidenthour = date[1]
        df_loc = pd.DataFrame({'id':[id_],'localisation':[localisation],'roadtype':[roadtype],'severitycategory':[severitycategory]\
                               ,'accidentyear':[accidentyear],'accidenttype':[accidenttype],'fsocommunecode':[fsocommunecode],\
                               'accidenttypecode':[accidenttypecode],'canton':[canton],'accidentday':[accidentday],\
                               'accidentmonth':[accidentmonth],'accidenthour':[accidenthour]})
        df  =df.append(df_loc,ignore_index=True)
        a+=1
    offset+=a
    if a==0:
        again = False
    
    
    
```

```python
import re
IS_DATE = re.compile("^[1-2]{1}[0-9]{3}$")
def describeDf(df,name="DESCRIPTION",level=1):
    print("______________________________ " + name+" ______________________________\n")
    date = []
    if(level==1):
        print("\n")
        for col in df.columns:
            if((not IS_DATE.match(col)) & (not re.compile("[0-9]+ an").match(col))):
                print("         ATTRIBUTE: "+col)
                print("   "+str(df[col].unique()))
                print("\n")
            else:
                date.append(col)
        print("         VALUES: "+str(date))
        print("\n\n\n")
    else :
        print(df.columns.values)
        print("\n\n\n")
```
