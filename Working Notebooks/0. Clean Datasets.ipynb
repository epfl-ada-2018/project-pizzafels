```python
# Functions
def propagationRow(df, feature,verbose=10000):
    for i in range(len(df[feature])):
        if i == 0:
            current = df[feature][i]
        else :
            try:
                if (isnan(float(df[feature][i]))):
                    df.loc[df.index == i, feature] = current
                else :
                    current = df[feature][i]
            except:
                current = df[feature][i]
            if(i%verbose == 0):
                print("[{}] - Rows {} to {} done".format(feature,i-verbose,i))

def get_X_first_columns(df,x):
    target = []
    if (x!=0):
        reg = re.compile("Unnamed: {}".format(x-1))
        for i in df.columns:
            target.append(i)
        if reg.match(i):
            return target
    else:
        reg = re.compile("Unnamed".format(x-1))
        for i in df.columns:
            if reg.match(i):
                target.append(i)
            else:
                return target
    
    

def completeDf(df, nb_cols):
    for col in get_X_first_columns(df,nb_cols):
        propagationRow(df,col,verbose=10000)
```

```python
FILE_NAME = 'Recettes_routes.xlsx'
ROW_TO_SKIP = 6
ID_LAST_ROW = None
TARGET_NAME = 'Recettes_routes'
NB_COLS_PROCEED = 0

IS_UNNAMED = re.compile("Unnamed")


target_noms = \
{'Unnamed: 0':'canton',\
'Unnamed: 1':'Total',\
'Total':'Total contributions',\
'Total.1':'Total taxes',\
'Adminis-':'Administration Prevention'}
```

```python
xls = pd.ExcelFile(data_folder+FILE_NAME)
s1 = xls.sheet_names
sheets = []
for s in s1:
    if int(s)>1991:
        sheets.append(s)
print (sheets)
```

```python
df = pd.read_excel(data_folder+FILE_NAME,skiprows=ROW_TO_SKIP,sheet_name ='2010',usecols='A:C,H').rename(columns=target_noms).dropna().reset_index()
for col in df.columns:
        if IS_UNNAMED.match(col):
            df = df.drop(columns=[col])
df
```

```python
merge = pd.DataFrame()
for sheet in sheets:
    df = pd.read_excel(data_folder+FILE_NAME,skiprows=ROW_TO_SKIP,sheet_name =sheet,usecols='A:C,H')
    df = df.rename(columns=target_noms)
    for col in df.columns:
        if IS_UNNAMED.match(col):
            df = df.drop(columns=[col])            
    df = df.dropna()
    df['annee'] = sheet
    merge = merge.append(df)
merge = merge.replace('...',-1)   
merge = merge.groupby(["annee","canton"]).sum()
```
