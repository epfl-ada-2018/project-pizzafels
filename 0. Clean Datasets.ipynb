```python
import os
import pandas as pd
import json
import folium
import matplotlib.pyplot as plt
import numpy as np
import re
import math
import numpy as np
from numpy import isnan
```

```python
# Functions
def propagationRow(df, feature,verbose=10000):
    for i in range(len(df[feature])):
        if i == 0:
            current = df[feature][i]
        else :
            try:
                if (isnan(float(df[feature][i]))):
                    df.loc[df.index == i, feature] = current
                else :
                    current = df[feature][i]
            except:
                current = df[feature][i]
            if(i%verbose == 0):
                print("[{}] - Rows {} to {} done".format(feature,i-verbose,i))

def get_X_first_columns(df,x):
    target = []
    if (x!=0):
        reg = re.compile("Unnamed: {}".format(x-1))
        for i in df.columns:
            target.append(i)
        if reg.match(i):
            return target
    else:
        reg = re.compile("Unnamed".format(x-1))
        for i in df.columns:
            if reg.match(i):
                target.append(i)
            else:
                return target
    
    

def completeDf(df, nb_cols):
    for col in get_X_first_columns(df,nb_cols):
        propagationRow(df,col,verbose=10000)
```

```python
df = pd.read_excel(data_folder + FILE_NAME + '.xlsx',skiprows=ROW_TO_SKIP,nrows=ID_LAST_ROW-ROW_TO_SKIP)
df.head(5)
```

```python
completeDf(df, NB_COLS_PROCEED)
df.head(5)
```
